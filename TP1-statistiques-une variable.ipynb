# Histogramme

Un histogramme sépare les valeurs possibles des données en classes ou groupes.

Pour chaque groupe, on construit un rectangle dont la base correspond aux valeurs de ce groupe et la hauteur correspond à l'effectif dans ce groupe.

Le code ci-dessous génère 3000 données aléatoires suivant une loi normale de moyenne 2000 et d'écart-type 200

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(2000, 200, 3000)
```

Représenter `data`sous forme d'histogramme et mettre les labels 'Salaires' et 'Effectifs' en abscisse et ordonnées. Le titre du graphique sera 'Répartition des salaires'.

```python
plt.xlabel("Salaires")
plt.ylabel("Effectifs")
plt.hist(data, bins =7)
```

Récupérer les bornes de chaque de chaque classe :

```python
plt.hist(data, bins = 8, color = 'yellow',
            edgecolor = 'red', align = 'mid')
```

Modifier ces bornes pour choisir [0, 1800, 2000, 2200, 4000] et afficher le nouvel histogramme :

```python
plt.hist(data, bins = [0, 1800, 2000, 2200, 4000], color = 'yellow',
            edgecolor = 'red', align = 'mid', )
```

## Boîte à moustaches

Appelées _box plot_ ou _whisker plot_ en anglais, elles sont un moyen de représenter le profil d'une série statistique quantitative.

Pour rappel, elles indiquent graphiquement : médiane, quartiles, minimum, maximum. 

Une boîte est créee entre le premier et le troisième quartile. Une ligne indique la médiane.

Reprendre les données de salaires et générer une _whisker plot_ :

```python
fig, axs = plt.subplots(1)
axs.boxplot(data)
plt.show()
```

Calculer le salaire moyen, médian ainsi que l'écart-type :

```python
print("moyenne :", np.average(data))
print("mediane :", np.median(data))
print("ecart-type :", np.std(data))
```

On affichera alors tous les partis politiques avec :

```python
with open('liste_deputes_excel.csv') as fichiercsv:
    reader = csv.DictReader(fichiercsv)
    for ligne in reader:
        print(ligne['Parti'], end=' ')
```

1) À l'aide de la structure `set` en python, donnez la liste des partis politiques présents dans le fichier. On utilisera la fonction `list()` et la liste sera nommée `liste_partis`

```python
with open('liste_deputes_excel.csv') as fichiercsv:
    reader = csv.DictReader(fichiercsv)
    liste_partis = set()  
    for ligne in reader:
        liste_partis.add(ligne['Parti'])
    print("\n")
    print(len(liste_partis))
    print(list(liste_partis))
```

2) Nous allons compter le nombre de députés par partis
* la structure de données dictionnaire semble la plus adaptée. Nous initialisons un dictionnaire et pouvons accéder à la valeur associée à une clé de la façon suivante :


```python
dico_partis = {parti:0 for parti in liste_partis}
print(dico_partis['LT'])


    
```

*  Ensuite comptez les députés par partis :

```python
with open('liste_deputes_excel.csv') as fichiercsv:
    reader = csv.DictReader(fichiercsv)
    for ligne in reader:
        dico_partis[ligne['Parti']] += 1
print(dico_partis)
```

3) Représenter sous forme de diagramme circulaire la répartition des députés à l'Assemble Nationale !

```python
tab1, tab2 = [], []
for i in dico_partis.keys():
    tab1.append(i)
for i in dico_partis.values():
    tab2.append(i)
tab3 = np.array(tab1)
y = np.array(tab2)

plt.pie(y, labels = tab1)
plt.show() 


```
